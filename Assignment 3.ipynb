```python
6. What happens to variables in a local scope when the function call returns?
```

Ans : When a function call returns, the variables defined in its local scope are no longer accessible and their memory is released by the program. However, if a variable is defined as a global variable, it will retain its value even after the function call returns. Additionally, if a variable from the local scope is passed as an argument to another function, its value may be retained in that function's scope.
